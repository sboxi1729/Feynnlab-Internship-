```python
# prompt: use k means

from sklearn.cluster import KMeans

# Select numerical columns for clustering
numerical_cols = ['time needed to visit in hrs', 'Google review rating', 'Entrance Fee in INR']
X = df1[numerical_cols]

# Determine the optimal number of clusters using the Elbow method
inertia = []
for i in range(1, 11):
  kmeans = KMeans(n_clusters=i, random_state=42)
  kmeans.fit(X)
  inertia.append(kmeans.inertia_)

plt.plot(range(1, 11), inertia, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.show()

# Based on the Elbow method, choose the optimal number of clusters (e.g., k=3)
k = 3
kmeans = KMeans(n_clusters=k, random_state=42)
df1['Cluster'] = kmeans.fit_predict(X)

# Visualize the clusters (example using scatter plot)
plt.scatter(df1['time needed to visit in hrs'], df1['Google review rating'], c=df1['Cluster'], cmap='viridis')
plt.title('Clusters of Tourist Places')
plt.xlabel('Time Needed to Visit (hrs)')
plt.ylabel('Google Review Rating')
plt.show()
```

```python
import string
def text_cleaning(text):
    text  = "".join([char for char in text if char not in string.punctuation])
    return text
```

```python
df2['Describe'] = df2['Describe'].apply(text_cleaning)
```

```python
df2.sample(3)
```

Reccomendation system ::

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from scipy.sparse import csr_matrix
from sklearn.neighbors import NearestNeighbors

tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(df2['Describe'])
tfidf_matrix.shape
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
cosine_sim
```

```python
indices = pd.Series(df2.index, index=df2['Name']).drop_duplicates()
indices
```

```python
def get_recommendations(title, cosine_sim=cosine_sim):

    idx = indices[title]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the scores of the 5 most similar food
    sim_scores = sim_scores[1:6]

    food_indices = [i[0] for i in sim_scores]
    return df2['Name'].iloc[food_indices]
```
